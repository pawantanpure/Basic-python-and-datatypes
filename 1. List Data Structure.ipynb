# 📋 Python Lists

A **list** is a built-in, mutable data structure used to store an ordered collection of items.

Lists are one of the most commonly used data structures in Python because they are flexible and easy to use.

# 💡 Quick Notes

- ✔ `append()` → Adds one item  
- ✔ `extend()` → Adds multiple items  
- ✔ `insert()` → Adds item at specific index  

- ✔ `remove()` → Removes by value  
- ✔ `pop()` → Removes by index  
- ✔ `clear()` → Removes all items  

- ✔ `sort()` → Sorts original list  
- ✔ `sorted()` → Returns a new sorted list  

- ✔ `len()` → Returns list length  
- ✔ `sum()` → Adds numeric values

````python
# List can contain any type or mixed type of data
l2 = [1,2,3,4,['a','b','c'],1+3j,34.7,True]
````

````python
# Copy list data it's not a shallow copy ( both are different address )
l3 = l2.copy()
# Check the memory address
print(id(l3))
print(id(l2))
````

````python
import copy 
l4 = copy.deepcopy(l2)
print(id(l4))
print(id(l2))
````
